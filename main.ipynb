```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('data.csv')
```

```python
data.head()
```

```python
data['month'] = pd.DatetimeIndex(data['date']).month
```

```python
data['year'] = pd.DatetimeIndex(data['date']).year
```

```python
data.head()
```

```python
# baseado na quantidade de likes, qual o top 10 palestras do TED talks
top_10 = data.sort_values(by=['likes'], ascending = False).head(10)
top_10
```

```python
# quais os 5 autores com mais vídeos no TED Talks
more_videos = data['author'].value_counts()
more_videos.head(5)
```

```python
# quantas palestras ocorreram em cada ano
total_year = data['year'].value_counts()
total_year
```

```python
# quantas palestras ocorreram em cada mês do ano de 2020
data_2022 = data[data['year'] == 2022]

ted_by_month = data.groupby(data_2022['month'])['month'].count()
ted_by_month
```
